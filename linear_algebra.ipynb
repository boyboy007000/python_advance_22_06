```python
def Hadamard_product(vector1, vector2):
    return (v1 * v2 for v1, v2 in zip(vector1, vector2))

def dor_product(vector1, vector2):
    return sum(v1 * v2 for v1, v2 in zip(vector1, vector2))
```

```python

def cosine_similarity(vector1, vector2):
    sumxy = sum([v1* v2 for v1, v2 in zip(vector1, vector2)])
    sumxx = sum([v1* v2 for v1, v2 in zip(vector1, vector1)])
    sumyy = sum([v1* v2 for v1, v2 in zip(vector2, vector2)])
    
    return sumxy /  (sumxx * sumyy)**0.5
```

```python
def cos_sim(img1, img2):
    
    img1 = img1.flatten()
    img2 = img2.flatten()
    
    img1 = img1.astype(np.float64)
    img2 = img2.astype(np.float64)
    
    #dot product img1, img2
    
    dot_img1 = np.dot(img1, img2)
    
    #norms img1, img2
    norm_img1 = np.linalg.norm(img1)
    norm_img2 = np.linalg.norm(img2)
    #simlaity
    return dot_img1 / (norm_img1 * norm_img2)

print(cos_sim(img1, img2).shape)
    
```

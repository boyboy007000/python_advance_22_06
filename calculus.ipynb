```python
def func(x, y):
    return x**2 + y**2

#tham so f là để gọi hàm của đồ thị ở đây là hàm func trên
def gradient(f, x, y, e = 1.04e-4):
    dx = (f(x+e/2, y) - f(x-e/2,y)) /e
    dy = (f(x, y +e/2 ) - f(x,y-e/2)) /e
    
    return dx, dy
```

```python
def func(a,b, data_x, data_y):
    
    error = []
    
    for i in range(len(data_x)):
        data_x_i = data_x[i]
        data_y_i = data_y[i]
        
        predict = a*data_x_i + b
        
        error_i = (data_y_i - predict)**2

        error.append(error_i)
    return sum(error)
```

```python
def gradient(func, a,b, data_x, data_y, e = 1.0e-4):
    f_a_1 = func(a+e/2, b, data_x, data_y)
    f_a_2 = func(a-e/2, b, data_x, data_y)
    da = (f_a_1 - f_a_2)/e
    
    f_b_1 = func(a, b+e/2, data_x, data_y)
    f_b_2 = func(a, b-e/2, data_x, data_y)
    db = (f_b_1 - f_b_2)/e
    
    return da,db
```

```python
data_x = [6.7,4.6,3.5,5.5]
data_y = [9.1,5.9,4.6,6.7]
a = -0.34
b = 0.04

da, db = gradient(func, a, b, data_x, data_y)
print(da, db)
```

```python
import random
import numpy as np

a = random.random()
b = 0.0

num_iterations = 300
step = 0.01

threshold = 1.0e-3

error_previous = 0
count = 0
for _ in range(num_iterations):
    da, db = gradient(func, a, b, data_x, data_y)
    
    sign_a = np.sign(da)
    sign_b = np.sign(db)
    
    a = a - sign_a*step
    b = b - sign_b*step
    
    error = func(a ,b, data_x, data_y)
    
    
    if (error - error_previous) < threshold:
        count+=1
    if count >10:    
        break
    
    error_previous = error
    print(a, b, error)
```
